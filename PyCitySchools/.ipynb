# Py City Schools Analysis # 

```python
import pandas as pd
```

```python
# Path to csv data files
MIN_PASSING_SCORE = 70
SCHOOL_DATA_PATH = "Resources/schools_complete.csv"
STUDENT_DATA_PATH = "Resources/students_complete.csv"

# Read the data files 
school_raw_df = pd.read_csv(SCHOOL_DATA_PATH)
school_raw_df
```

```python
student_raw_df = pd.read_csv(STUDENT_DATA_PATH)
student_raw_df
```

```python
# Combine the data into a single data set 
complete_df = pd.merge(school_raw_df, student_raw_df, how="left", on=["school_name", "school_name"])
complete_df
```

# District Summary #

```python
# Total number of unique schools
school_count = complete_df["school_name"].value_counts()
total_school_count = len(school_count)
total_school_count
```

```python
# Total number of students
student_count = complete_df["Student ID"].value_counts()
total_student_count = len(student_count)
total_student_count
```

```python
# Total budget 
total_budget = school_raw_df["budget"].sum()
total_budget
```

```python
# Average math score
avg_math_score = complete_df["math_score"].mean()
avg_math_score
```

```python
# Average reading score
avg_reading_score = complete_df["reading_score"].mean()
avg_reading_score
```

```python
# % passing math (the percentage of students who passed math)
passing_math_s = complete_df["math_score"] >= MIN_PASSING_SCORE
print(passing_math_s)
students_passing_math = complete_df.loc[passing_math_s]
num_student_passing_math = students_passing_math["Student ID"].count()

percent_passing_math = (num_student_passing_math / total_student_count) * 100
percent_passing_math
```

```python
# % passing reading (the percentage of students who passed reading)
passing_reading_s = complete_df["reading_score"] >= MIN_PASSING_SCORE
students_passing_reading = complete_df[passing_reading_s]
num_student_passing_reading = students_passing_reading["Student ID"].count()

percent_passing_reading = (num_student_passing_reading / total_student_count) * 100
percent_passing_reading
```

```python
# % overall passing (the percentage of students who passed math AND reading)
combined_passing_count = complete_df[passing_math_s & passing_reading_s]['Student ID'].count()
combined_passing_pct = 100 * (combined_passing_count / total_student_count)
combined_passing_pct
```

```python
# Create a dataframe to hold the results
district_summary_df = pd.DataFrame({
    "Total Schools": total_school_count,
    "Total Students": f"{total_student_count:,}",
    "Total Budget": f"${total_budget:,.2f}",
    "Average Math Score": f"{avg_math_score:.6f}",
    "Average Reading Score": f"{avg_reading_score:.5f}",
    "% Passing Math": f"{percent_passing_math:.6f}",
    "% Passing Reading": f"{percent_passing_reading:.6f}",
    "% Overall Passing": f"{combined_passing_pct: .6f}"
}, index=[0])

district_summary_df
```

# School Summary #

```python
school_raw_df
```

```python
# Reser the pandas index in a new dataframe
school_type = school_raw_df.set_index(["school_name"])
school_type
```

```python
# Use the code provided to select the school type - get a series for all school types
school_types = school_type["type"]
school_types
```

```python
# Calculate the per school student count
per_school_student_counts_s = student_raw_df["school_name"].value_counts()
per_school_student_counts_s


#student_counts_sum = per_school_student_counts_s.sum()
#student_counts_sum
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = complete_df.groupby(["school_name"]).mean()["budget"]
print(per_school_budget)

per_school_capita = per_school_budget / per_school_student_counts_s
per_school_capita
```

```python
# Calculate the average test scores
per_school_math_s = student_raw_df.groupby("school_name")["math_score"].mean()
print(per_school_math_s)


per_school_reading_s = student_raw_df.groupby("school_name")["reading_score"].mean()
per_school_reading_s
```

```python
# Calculate the number of schools with math scores of 70 or higher 
school_passing_math_df = complete_df[complete_df["math_score"] >= 70]
school_passing_math_df
```

```python
# Calculate the number of schools with reading scores of 70 or higher 
school_passing_reading_df = complete_df[complete_df["reading_score"] >= 70]
school_passing_reading_df
```

```python
# students passing reading (if needed)
school_passing_reading_df = student_raw_df[student_raw_df["reading_score"] >= 70]
school_passing_reading_df
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
school_passing_df = complete_df[(complete_df["math_score"] >= 70) & (complete_df["reading_score"] >= 70)]
school_passing_df                                      
```

```python
# Use a groupby to count the number of students at each school who passed math
school_passing_math_df = school_passing_math_df.groupby("school_name")
per_school_passing_math = school_passing_math_df["student_name"].count()
per_school_passing_math
```

```python
# Use a groupby to count the number of students at each school who passed reading
school_passing_reading_df = school_passing_reading_df.groupby("school_name")
per_school_passing_reading = school_passing_reading_df["student_name"].count()
per_school_passing_reading
```

```python
# Students who passed both math and reading
school_passing_df = school_passing_df.groupby("school_name")
passing_students_per_school = school_passing_df["student_name"].count()
passing_students_per_school
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math_pct = (per_school_passing_math / per_school_student_counts_s) * 100
per_school_passing_math_pct

per_school_passing_reading_pct = (per_school_passing_reading / per_school_student_counts_s) * 100
per_school_passing_reading_pct

overall_passing_pct = (passing_students_per_school / per_school_student_counts_s) * 100
overall_passing_pct

```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary_df = pd.DataFrame({
            "School Types": school_types,
            "Total Students": per_school_student_counts_s,
            "Total School Budget": per_school_budget,
            "Per Student Budget": per_school_capita,
            "Average Math Test Scores": per_school_math_s,
            "Average Reading Test Scores": per_school_reading_s,
            "% Passing Math": per_school_passing_math_pct,
            "% Passing Reading": per_school_passing_reading_pct, 
            "Overall Passing Rate": overall_passing_pct
    
})

# Formatting
per_school_summary_df["Total Students"] = per_school_summary_df["Total Students"].map("{:,}".format)
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)
per_school_summary_df["Average Math Test Scores"] = per_school_summary_df["Average Math Test Scores"].map("{:.2f}%".format)
per_school_summary_df["Average Reading Test Scores"] = per_school_summary_df["Average Reading Test Scores"].map("{:.2f}%".format)
per_school_summary_df["% Passing Math"] = per_school_summary_df["% Passing Math"].map("{:.2f}%".format)
per_school_summary_df["% Passing Reading"] = per_school_summary_df["% Passing Reading"].map("{:.2f}%".format)
per_school_summary_df["Overall Passing %"] = per_school_summary_df["Overall Passing Rate"].map("{:.2f}%".format)


# Display the DataFrame
per_school_summary_df
```

# Scores by School Spending

```python
# Establish the bins

size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# copy
school_spending_df = per_school_summary_df.copy()
school_spending_df
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, size_bins, labels=labels)
school_spending_df
```

```python

school_spending_df["Average Math Test Scores"] = school_spending_df["Average Math Test Scores"].str.replace("%", "").astype(float)
school_spending_df["Average Reading Test Scores"] = school_spending_df["Average Reading Test Scores"].str.replace("%", "").astype(float)
school_spending_df["% Passing Math"] = school_spending_df["% Passing Math"].str.replace("%", "").astype(float)
school_spending_df["% Passing Reading"] = school_spending_df["% Passing Reading"].str.replace("%", "").astype(float)
school_spending_df["Overall Passing %"] = school_spending_df["Overall Passing %"].str.replace("%", "").astype(float)

school_spending_df
```

```python
# first, groupby the spending ranges
spending_math_df = school_spending_df.groupby("Spending Ranges (Per Student)")
spending_grouped_df = spending_math_df.mean()

spending_math_series = spending_grouped_df["Average Math Test Scores"]
spending_reading_series = spending_grouped_df["Average Reading Test Scores"]
spending_passing_math_series = spending_grouped_df["% Passing Math"]
spending_passing_reading_series = spending_grouped_df["% Passing Reading"]
spending_passing_series = spending_grouped_df["Overall Passing %"]
```

```python
# Assemble into Dataframe

spending_summary_df = pd.DataFrame(
    {
        "Average Math Test Scores" : spending_math_series,
        "Average Reading Test Scores" : spending_reading_series,
        "% Passing Math" : spending_passing_math_series,
        "% Passing Reading" : spending_passing_reading_series,
        "Overall Passing %" : spending_passing_series
    }
)

spending_summary_df["Average Math Test Scores"] = spending_summary_df["Average Math Test Scores"].map("{:.2f}".format)
spending_summary_df["Average Reading Test Scores"] = spending_summary_df["Average Reading Test Scores"].map("{:.2f}".format)
spending_summary_df["% Passing Math"] = spending_summary_df["% Passing Math"].map("{:.2f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:.2f}".format)
spending_summary_df["Overall Passing %"] = spending_summary_df["Overall Passing %"].map("{:.2f}".format)

spending_summary_df

```
